```python
data1=data.drop(['lat','lon'],axis=1) #drop the unnecessary variables and axis=1 represents columns
data1
```

```python
data2=data1.drop('ID',axis=1) # no need to use list for removing one variable
data2
```

```python
x=data2.drop('price',axis=1) # place all independent variables in one variable
y=data2['price']  #place the target variable in another variable (dependent variables)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0) #it ensures that the same randomization is used each time you run the code, resulting in the same splits of the data.
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_test
```

```python
reg.predict(x_test)
```

```python
ypred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score #gives us the accuracy of the model
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,ypred)  # gives us the units of error
```

```python
results=pd.DataFrame(columns=['Price','Predicted'])
results['Price']=y_test
results['Predicted']=ypred
results=results.reset_index()
results['ID']=results.index
results
```

```python
#plotting a line plot to showcase the difference between actual and predicted values
sns.lineplot(x='ID',y='Price',data=results.head(50),label="actual")
sns.lineplot(x='ID',y='Predicted',data=results.head(50),label="predicted")
```
